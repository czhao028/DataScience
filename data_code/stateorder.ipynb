```python
state_orders = pd.read_csv(r'../excel_data\state_wide_orders.csv')
state_orders.set_index('State abbr', inplace=True)
```

```python

status_2 = state_orders[state_orders['Status'] == 2]
status_2.drop(columns=['STATE', 'State', 'Status'], inplace=True)
status_2
```

```python
status_2_dict = {}
for ind in status_2.index:
    status_2_dict[ind] = status_2.loc[ind].dropna().as_matrix()
    
status_2_dict
    
```

```python
def statewide_order_county(row, s2, allstates):
    county_state = row['display_name']
    print(county_state)
    state = county_state.split(", ")[1]
    county = county_state.split(", ")[0]
    status = allstates.loc[state]['Status']
    if status == 1 or status == 0:
        return status
    else:
        if county in s2[state]:
            return 1
        return 0

x = statewide_order_county(df.loc[1], status_2_dict, state_orders)
x
```

```python
df['statewide_order'] = df.apply(lambda row: statewide_order_county(row, status_2_dict, state_orders), axis=1)
```
